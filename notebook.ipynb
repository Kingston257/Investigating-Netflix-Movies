```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
netflix_df = pd.read_csv("netflix_data.csv") #load netflix data
print(netflix_df[0:5]) #inspect data using the first 5 rows

#subset the DataFrame for "Movie"
netflix_subset = netflix_df[netflix_df["type"]=="Movie"]
print(netflix_subset)
type(netflix_subset)
```

```python
#Subset netflix_subset into netflix_movies
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
print(netflix_movies)
```

```python
#filter the netflix_movies dataframe by movie duration
short_movies = netflix_movies[netflix_movies["duration"] < 60]

fig = plt.figure(figsize=(12, 8))
plt.scatter(short_movies['release_year'], short_movies['duration'])
#plt.xlabel('Release year')
#plt.ylabel('Duration (min)')
plt.title('Short Movies Duration by Year of Release')
plt.show()
```

```python
#first 20 rows of short_movies
print(short_movies[0:20])
```

```python
#assign colors to movie genres
colors = []
color_map = {
    'Children': 'yellow',
    'Documentaries': 'green',
    'Stand-Up': 'blue',
    'others': 'gray'
    
}
for value in netflix_movies['genre']:
    if value in color_map:
        colors.append(color_map[value])
    else:
        colors.append(color_map['others'])
```

```python
#plot the movie duration by year using the genre colors created
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors) # specify a colormap for the colors
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')
plt.show()
```
